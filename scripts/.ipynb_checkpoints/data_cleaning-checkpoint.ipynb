# Read in the WHO data and clean it

```python
# read in who data
who_data = pd.read_csv("data/raw/WHO_COVID_19_global_daily_data.csv")
who_data
```

```python
# filter for Canada
who_filtered = who_data[who_data['Country'] == 'Canada']
who_filtered
```

```python
# filter for the years we want
who_filtered['Date_reported'] = pd.to_datetime(who_filtered['Date_reported'])  # Convert to datetime

who_filtered_new = who_filtered[who_filtered['Date_reported'].dt.year.isin([2020, 2022])]
who_filtered_new
```

```python
# replace NAs in 2020 with zeros - from cumulative cases column we know there are no new cases
who_filtered_new.loc[who_filtered_new['Date_reported'].dt.year == 2020] = who_filtered_new.loc[who_filtered_new['Date_reported'].dt.year == 2020].fillna(0)
who_filtered_new
```

```python
# summarize by week
# filter for 2020 and 2022 before June 12
filtered_df = who_filtered_new[
    (who_filtered_new['Date_reported'].dt.year.isin([2020, 2022])) &
    (who_filtered_new['Date_reported'].dt.month <= 6) & 
    (who_filtered_new['Date_reported'].dt.day < 12)
]

# group by week and sum numerical columns
weekly_summary = filtered_df.resample('W-Mon', on='Date_reported').sum()

# reset index to make Date_reported a column again
weekly_summary = weekly_summary.reset_index()
weekly_summary
```

```python
# summarize 2020 by week because 2022 is by week
filtered_df = who_filtered_new[(who_filtered_new['Date_reported'] > '2022-06-11') & (who_filtered_new['New_cases'].notna())]

# grooup the filtered data by week and sum the numerical columns
weekly_summary_new = filtered_df.groupby(pd.Grouper(key='Date_reported', freq='W-Mon')).agg({
    'New_cases': 'sum',
    'New_deaths': 'sum',
    'Cumulative_cases': 'first'  # Keep the first Cumulative_cases for the week
}).reset_index()

# reset the index to make Date_reported a column again
weekly_summary_new = weekly_summary_new.reset_index()

# put these together
weekly_summary = pd.concat([weekly_summary, weekly_summary_new])

weekly_summary
```

```python
# select certain columns
weekly_summary = weekly_summary[['Date_reported', 'New_cases', 'Cumulative_cases']]
weekly_summary['year'] = weekly_summary['Date_reported'].dt.year  # Extract the year
weekly_summary = weekly_summary.loc[~weekly_summary['year'].isin([2021, 2023])]
weekly_summary
```

```python
# add row for 2018
new_row = pd.DataFrame({'Date_reported': ['2018-11-11'], 'New_cases': [0], 'Cumulative_cases': [0], 'year': [2018]})

# add the new row to the existing df
weekly_summary = pd.concat([weekly_summary, new_row], ignore_index=True)
weekly_summary
```

```python
# now summarize the average by year
avg_new_cases_by_year = weekly_summary.groupby('year')['New_cases'].mean()
avg_new_cases_by_year
```

# Merge the two datasets together

```python
# now merge the two datasets together
merged_data = pd.merge(merged_data, avg_new_cases_by_year, on=['year'], how='outer')
merged_data
```

```python
# write the new clean file
merged_data.to_csv('data/merged_data_clean.csv', index=False)
```
