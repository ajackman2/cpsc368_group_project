```python
internet_activities = ['access_to_internet', 'sent_received_emails', 'instant_messaging', 'social_media', 'voice_calls']
df_melted = pd.melt(df, 
                    id_vars=['age_group', 'year', 'New_cases'], 
                    value_vars=internet_activities,
                    var_name='activity_type', 
                    value_name='percentage')
df_melted
```

```python
chart = alt.Chart(df_melted).mark_line(point=True).encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('percentage:Q', title='Percentage of Population (%)'),
    color=alt.Color('age_group:N', title='Age Group'),
    column=alt.Column('activity_type:N', title='Internet Activity'),
    tooltip=['year:O', 'age_group:N', 'percentage:Q', 'New_cases:Q']
).properties(
    width=175,
    height=300,
    title='Changes in Internet Usage Across Age Groups (2018-2022)'
)
chart
```

##### Initial Observations 
 - Significant increase in usage of internet services in Seniors(>65 years).
 - Stagnated change in voice calls after 2020.

```python
df_2018 = df[df['year'] == 2018].copy()
df_2020 = df[df['year'] == 2020].copy()
df_2022 = df[df['year'] == 2022].copy()
```

```python
results = []

for age in df_2018['age_group'].unique():
    for activity in internet_activities:
        val_2018 = df_2018[df_2018['age_group'] == age][activity].values[0]
        val_2020 = df_2020[df_2020['age_group'] == age][activity].values[0]
        val_2022 = df_2022[df_2022['age_group'] == age][activity].values[0]
        
        # Percentage change
        change_2018_2020 = ((val_2020 - val_2018) / val_2018) * 100
        change_2020_2022 = ((val_2022 - val_2020) / val_2020) * 100
        
        results.append({
            'age_group': age,
            'activity': activity,
            'period': '2018-2020 (Early Pandemic)',
            'percent_change': change_2018_2020,
            'covid_cases': df_2020[df_2020['age_group'] == age]['New_cases'].values[0]
        })
        
        results.append({
            'age_group': age,
            'activity': activity,
            'period': '2020-2022 (Later Pandemic)',
            'percent_change': change_2020_2022,
            'covid_cases': df_2022[df_2022['age_group'] == age]['New_cases'].values[0]
        })

df_changes = pd.DataFrame(results)
df_changes
```

```python
heatmap = alt.Chart(df_changes).mark_rect().encode(
    x=alt.X('activity:N', title='Internet Activity'),
    y=alt.Y('age_group:N', title='Age Group', sort=['15 to 24 years', '25 to 44 years', '45 to 64 years', '65 years and over']),
    color=alt.Color('percent_change:Q', scale=alt.Scale(scheme='blueorange', domain=[-10, 40]), title='% Change'),
    column=alt.Column('period:N', title='Time Period'),
    tooltip=['age_group:N', 'activity:N', 'percent_change:Q', 'covid_cases:Q']
).properties(
    width=150,
    height=100,
    title='Impact of Pandemic on Internet Usage Patterns by Age Group'
)
heatmap
```

```python
text = alt.Chart(df_changes).mark_text(color='black', fontSize=9).encode(
    x=alt.X('activity:N'),
    y=alt.Y('age_group:N', sort=['15 to 24 years', '25 to 44 years', '45 to 64 years', '65 years and over']),
    text=alt.Text('percent_change:Q', format='.1f'),
    column=alt.Column('period:N')
)
text
```

##### Visualization Submission
The visualization created shows the impact of the pandemic on internet usage patterns across different age groups from 2018 to 2022. It displays:

 1.  A line chart showing trends in internet activities across years(2018-2022)
 2.  A heatmap showing percentage changes in internet activities during the early pandemic (2018-2020) and later pandemic periods (2020-2022)
 3.  Text annotations showing the exact percentage change for each activity and age group

##### Purpose and Suitability of the Visualization
This visualization is meant to examine:

1. How has Canadians' daily internet use changed with the prevalence of COVID-19 cases?
2. Which internet activities saw the most significant changes and whether certain age groups were more affected than others

The heatmap is particularly suitable for this analysis because:

- It allows for easy comparison across multiple variables (age groups, activities, time periods)
- Color intensity immediately highlights the most significant changes
- The percentage values provide precise quantitative information
- The column layout separates early pandemic effects from later pandemic trends
